```python
data=pd.read_csv('/kaggle/input/mushroom-attributes/mushroom.csv')
```

```python
data.head()
```

```python
cols=data.columns.values.tolist()
```

```python
d=[]
for col in cols:
    unq=data[col].unique()
    tmp={}
    for j in range(len(unq)):
        tmp[unq[j]]=j
    data[col]=data[col].map(tmp)
data.head()
```

```python
data.info()
```

```python
from keras.models import Sequential

model=Sequential()
```

```python
import keras
```

```python
model.add(keras.layers.Dense(23, activation="relu"))
model.add(keras.layers.Dense(3, activation="relu"))
model.add(keras.layers.Dense(1))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
X=data.drop(['class'],axis=1)
```

```python
y=data['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2,  random_state=32)
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=40,validation_data=(X_test,y_test))
```
